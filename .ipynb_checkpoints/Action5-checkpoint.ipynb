# Ví dụ xích Markov không thuần nhất
Ta giả sử có hai trạng thái A và B có các phân bố $\pi(A) = 2/5, \pi(B) = 3/5$ và có ma trận chuyển trạng thái như sau:  
\begin{align*}
P = 
\begin{pmatrix}
1/4 & 3/4 \\
1/2 & 1/2
\end{pmatrix}
\end{align*}
Nếu $P(X_0 = A) = 2/5$ và $P(X_0 = B) = 3/5$, ta được:
\begin{align*}
P(X_1=A)&= P(X_1=A|X_0=A)P(X_0=A)+P(X_1=A|X_0=B)P(X_0=B)\\
        &=1/4.2/5+1/2.3/5=2/5 \\
\implies P(X_1=B)&=1-P(X_1=A)=3/5
\end{align*}
Ta thấy $X_1$ có cùng phân phối với $X_0$, nếu tiếp tục tính toán ta cũng thấy $X_n$ có cùng phân phối với $X_0$. Ta nói đây là xích Markov thuần nhất.  
Tuy nhiên nếu phân phối ban đầu $P(X_0=A)=1$ thì $P(X_1=A)=1/4$ và do đó $X_1$ không có cùng phân phối với $X_0$. Đây là xích Markov không thuần nhất.

# Phát biểu và chứng minh phương trình Chapman-Kolmogorov
Phương trình Chapman-Kolmogorov: Với mọi $m,n\geq0$ và $i,j\in E$, ta có:
\begin{align*}
p_{ij}^{(m+n)}=\sum_{r\in E}p_{ir}^{(m)}p_{rj}^{(n)}.
\end{align*}
với $p_{ij}^{(m+n)}$ là xác suất chuyển từ trạng thái $i$ sang trạng thái $j$ qua $m+n$ bước chuyển.  

#### Chứng minh:
Theo định nghĩa xác suất có điều kiện, ta có:
\begin{align*}
p_{ij}^{(m+n)}&=\sum_{r\in E}P(X_{m+n}=j,X_m=r,X_0=i)/P(X_0=j) \\
            &=\sum_{r\in E}P(X_{m+n}=j|X_m=r,X_0=i)P(X_m,X_0=i/P(X_0=j) \\
            &=\sum_{r\in E}p_{ir}^{(m)}p_{rj}^{(n)}
\end{align*}

#### Nhận xét:
Nếu $P^{(n)}$ là ma trận xác suất chuyển qua n bước chuyển $P_{ij}^n$, theo phương trình Chapman-Kolmogorov ta được:
\begin{align*}
P^{(n)} = P^{(n-1+1)} = P^{n-1}.P = P^n
\end{align*}

# Ví dụ
Giả sử rằng khả năng ngày mai trời mưa phụ thuộc vào thời tiết ngày hôm nay có mưa hay không. Nếu hôm nay trời mưa thì khả năng ngày mai trời mưa là $0.7$; và nếu hôm nay trời nắng thì khả năng ngày mai trời mưa là $0.4$. Với $0$ là trời mưa và $1$ là trời nắng, tính xác suất trời sẽ mưa 4 ngày sau, biết rằng hôm nay trời mưa.
#### Giải
Ta có ma trận xác suất chuyển:
\begin{equation*}
P =
\begin{pmatrix}
0.7 & 0.3 \\
0.4 & 0.6
\end{pmatrix}
\end{equation*}
Ta áp dụng Chapman-kolmogorov:
\begin{align*}
P^{(2)} = P^2 &=
\begin{pmatrix}
0.7 & 0.3 \\
0.4 & 0.6
\end{pmatrix}
.
\begin{pmatrix}
0.7 & 0.3 \\
0.4 & 0.6
\end{pmatrix}
\\ &=
\begin{pmatrix}
0.61 & 0.39 \\
0.52 & 0.48
\end{pmatrix}
\\ P^{(4)} = (P^2)^2 &=
\begin{pmatrix}
0.61 & 0.39 \\
0.52 & 0.48
\end{pmatrix}
.
\begin{pmatrix}
0.61 & 0.39 \\
0.52 & 0.48
\end{pmatrix}
\\ &=
\begin{pmatrix}
0.5749 & 0.4251 \\
0.5668 & 0.4332
\end{pmatrix}
\end{align*}
Vậy xác suất cần tìm là $P_{00}^4=0.5749$
