```python
import numpy as np
from graphviz import Digraph
```

```python
states = ['Sleep','Run','Icecream']
P = [[0.2,0.6,0.2],[0.1,0.6,0.3],[0.2,0.7,0.1]]
n = 4
```

```python
if sum([sumP[i]) for i in range(len(P))]) != len(P):
        print("Wrong Transition Matrix")
else:
    n_step_P = P
    for i in range(n-1):
        n_step_P = np.dot(n_step_P, P)
n_step_P = np.around(n_step_P,2)
print(n_step_P)
```

```python
G = Digraph()
G.attr(rankdir='LR', size='20')
G.attr('node',shape='circle')

for i in states:
    G.node(i)
    
for i in range(len(P)):
    for j in range(len(P)):
        if (P[i][j]!=0):
            G.edge(states[i],states[j],label=str(P[i][j]))
G
```
