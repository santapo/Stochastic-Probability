```python
import numpy as np

def computeK(n,P):
    A = np.copy(P)
    count = 0
    
    for i in range(n):
        if (P[i,i]==1):
            A = np.delete(A,i,0)
            A = np.delete(A,i,1)
            count += 1
            
    A = A - np.identity(n-count)
    
    return np.linalg.solve(A,-np.ones(n-count))
```
