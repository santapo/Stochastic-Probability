#### ***Câu 1:***
\begin{align*}
P = 
    \begin{bmatrix}
    3/7 & 3/7 & 1/7 \\
    0.23 & 0.47 & 0.3 \\
    1/11 & 2/11 & 8/11
    \end{bmatrix}
\end{align*}

Ta có:
\begin{equation}
Y_n = 
    \begin{cases}
    1 & \mbox{nếu } X_n = 0 \\
    -1 & \mbox{nếu } X_n \neq 0
    \end{cases}
\end{equation}

Để chứng minh $(Y_n)_{n\geq 0}$ là xích Markov ta chứng minh nó có tính Markov.  
Vì $(X_n)_{n\geq 0}$ là xích Markov nên $Y_n$ độc lập với $Y_1,Y_2...Y_{n-2}$.  

\begin{align*}
P(Y_n|Y_{n-1}...Y_0) &= \frac{P(Y_n,Y_{n-1}...Y_0)}{P(Y_{n-1}...Y_0)} \\
                    &= \frac{P(Y_n,Y_{n-1})...P(Y_0)}{P(Y_{n-1})...P(Y_0)} \\
                    &= \frac{P(Y_n,Y_{n-1})...P(Y_0)}{P(Y_{n-1})}
\end{align*}

Vậy $(Y_n)_{n\geq 0}$ có tính Markov nên nó là xích Markov.  
Dễ dàng tính được:

\begin{align*}
P(Y_{n+1}=-1|P_n=-1) &= \frac{P(X_{n+1}=0|X_n=0)}{\sum_{i \in \mathbb{I}}P(X_{n+1}=0|X_n=i)}= \frac{3}{7} \\
\\
P(Y_{n+1}=-1|P_n=1) &= \frac{\sum_{i \in \{1,2\}}P(X_{n+1}=0|X_n=i)}{\sum_{j \in \mathbb{I}}\sum_{i \in \{1,2\}}P(X_{n+1}=j|X_n=i)}= 0.161
\end{align*}

    \end{bmatrix}
\end{align*}

#### ***Câu 4:***

```python
def transitiveClosure(n,graph,start,end): 
        
        reach =[i[:] for i in graph] 

        for k in range(n): 
            for i in range(n): 
                for j in range(n): 
                    reach[i][j] = reach[i][j] or (reach[i][k] and reach[k][j])
        return reach[start][end]
def isConnected(n,graph,i,j):
    if transitiveClosure(n,graph,i,j) and transitiveClosure(n,graph,j,i):
        print('i <---> j')
    else:
        print('Unconnected')
def isReachable(n,graph,i,j):
    if transitiveClosure(n,graph,i,j):
        print('i ---> j')
    else:
        print('Unreachable')
```
