    \end{bmatrix}
\end{align*}
Ta đi tính $P_{11}^{(n)}$  
Áp dụng phương trình Chapman-Kolmogorov, ta có:

\begin{align*}
P_{11}^{(n+1)} = P_{12}^{(n)}(1-\beta)+P_{11}^{(n)}\alpha
\end{align*}
Mà $P_{11}^{(n)} + P_{12}^{(n)} = 1$ nên ta thu được phương trình truy hồi cho $P_{11}^{(n+1)}$ như sau:

\begin{align*}
P_{11}^{(n+1)} = (1-\alpha -\beta)P_{12}^{(n)}+1-\beta, \space P_{11}^{(0)} = 1
\end{align*}
Từ đây suy ra

\begin{equation}
P_{11}^{(n)} = 
\begin{cases}
\frac{\beta-1}{\alpha+\beta-2}+\frac{\alpha-1}{\alpha+\beta-2}(\alpha+\beta-1)^n & \mbox{nếu } \alpha + \beta >0 \\
1 & \mbox{nếu } \alpha + \beta = 0
\end{cases}
\end{equation}
Tương tự với $P_{12}^{(n)}, P_{21}^{(n)}$và $P_{22}^{(n)}$, ta được:

\begin{equation}

Mặt khác:

\begin{align*}
P(Y_n|Y_{n-1}..Y_0) &= P(X_{n+1},X_n|X_n,X_{n-1}..X_0) \\

                    &= P(Y_{n+1}|Y_n)
\end{align*}
Suy ra $Y_n$ có tính Markov. Vậy $Y_n$ là một  xích Markov

#### ***Câu 5:***
*(a)* $Y_n=X_n+X_{n+1}$

\begin{align*}
P(Y_n|Y_{n-1}..Y_0) &= P(X_{n+1}+X_n|X_n+X_{n-1},X_{n-1}+X_{n-2}..X_0+X_1)
\end{align*}

Mà $X_0+X_1,X_1+X_2...X_{n-1} + X_n$ không độc lập với nhau hay là

\begin{align*}
P(X_{n+1}+X_n|X_n+X_{n-1},X_{n-1}+X_{n-2}..X_0+X_1) \neq P(X_{n+1}+X_n|X_n+X_{n-1})
\end{align*}

Suy ra $Y_n$ không có tính Markov nên không phải xích Markov

*(b)* $Z_n = X_n\times X_{n+1}$
Vì $X_0,X_1..X_n$ đọc lập với nhau nên $X_n\times X_{n+1} = \emptyset$ hay là $Z_n$ không tồn tại phân phối ban đầu (không có trạng thái)  
Vậy $Z_n$ không phải xích Markov
